```python
sum=0
count=0
df=df.fillna(0)
cf=df['Laid_Off_Count'].tolist()
ui=np.array(cf)
# convert each string element to a number using a list comprehension
arr_numbers = [float(x) for x in ui]  # use int(x) for integers
xi=ui.astype(int)
for x in arr_numbers:
    if isinstance(x, (int, float)):
        sum += int(x)
        count += 1

mean = sum / count
```
